```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
plt.style.use('ggplot')
```

```python
dataDir='./fep-3lipid-pore/'
filepath='/'.join([dataDir,'fep.txt'])
lipid=pd.read_table(filepath,header=None,delim_whitespace=True)
lipid.columns=['Time','FEP','mbar','std']

plt.rcParams['axes.facecolor']='white'
import matplotlib
matplotlib.rc('axes',edgecolor='k')

fig,ax1 = plt.subplots(figsize=(18,6))
ax1.scatter(lipid['Time'],lipid['FEP'],s=100,color='k')
ax1.plot(lipid['Time'],lipid['FEP'],linewidth=5,color='k')

#ax1.legend(fontsize=20)
ax1.set_xlim(0,2.4)
ax1.set_xticks(np.arange(0,2.4,0.4))
ax1.tick_params(axis='x', colors='black')
ax1.tick_params(axis='y', colors='black')
ax1.tick_params(axis='y', which='major', labelsize=30)
ax1.tick_params(axis='x', which='major', labelsize=30)
ax1.set_xlabel('Time (ns)',fontsize=30,color='k')
ax1.set_ylabel('FEP (kcal/mol)',fontsize=30,color='k')
```

```python
dataDir='./fep-3lipid-pore'
filepath='/'.join([dataDir,'fep_bulk40.txt'])
lipid=pd.read_table(filepath,header=None,delim_whitespace=True)
lipid.columns=['Time','FEP40']

filepath='/'.join([dataDir,'fep_bulk60.txt'])
lipid1=pd.read_table(filepath,header=None,delim_whitespace=True)
lipid1.columns=['Time','FEP60']


plt.rcParams['axes.facecolor']='white'
import matplotlib
matplotlib.rc('axes',edgecolor='k')

fig,ax1 = plt.subplots(figsize=(18,6))
ax1.scatter(lipid['Time'],lipid['FEP40'],s=100,color='k')
ax1.plot(lipid['Time'],lipid['FEP40'],linewidth=5,color='k')
ax1.plot(lipid1['Time'],lipid1['FEP60'],linewidth=5,color='g')
ax1.scatter(lipid1['Time'],lipid1['FEP60'],s=100,color='g')
ax1.legend(fontsize=20)
ax1.set_xlim(0,12)
ax1.set_xticks(np.arange(0,13,1))
ax1.tick_params(axis='x', colors='black')
ax1.tick_params(axis='y', colors='black')
ax1.tick_params(axis='y', which='major', labelsize=30)
ax1.tick_params(axis='x', which='major', labelsize=30)
ax1.set_xlabel('Time (ns)',fontsize=30,color='k')
ax1.set_ylabel('FEP ($kcal/mol$)',fontsize=30,color='k')
ax1.set_ylim(-80,-60)
ax1.set_yticks(np.arange(-80,-55,5))
```

```python
dataDir='./fep-3lipid-pore/'
filepath='/'.join([dataDir,'final.uncorr.job3.txt'])
lipidun=pd.read_table(filepath,header=None,delim_whitespace=True)
lipidun.columns=['RepID','UncorrNum']

plt.rcParams['axes.facecolor']='white'
import matplotlib
matplotlib.rc('axes',edgecolor='k')

fig,ax1 = plt.subplots(figsize=(18,4))
ax1.bar(lipidun['RepID'],lipidun['UncorrNum'],alpha=1,color='g')
ax1.set_ylabel(r"3lipids-in-pore",color="black",fontsize=20)
ax1.tick_params(axis='x', colors='black')
ax1.tick_params(axis='y', colors='black')
ax1.tick_params(axis='y', which='major', labelsize=20)
ax1.tick_params(axis='x', which='major', labelsize=20)
ax1.set_xlim(0,128)
ax1.set_xticks(np.arange(0,130,10))
ax1.set_ylim(0,1300)
ax1.set_yticks(np.arange(0,1400,200))

print np.min(lipidun['UncorrNum'])
print np.max(lipidun['UncorrNum'])
```

```python
dataDir='./fep-3lipid-pore/'
filepath='/'.join([dataDir,'final.acc.txt'])
lipidRatio=pd.read_table(filepath,header=None,delim_whitespace=True)
lipidRatio.columns=['RepID1','RepID2','AccR']

plt.rcParams['axes.facecolor']='white'
import matplotlib
matplotlib.rc('axes',edgecolor='k')

width=0.7

plt.subplots(figsize=(20,4))
plt.bar(lipidRatio.RepID1, lipidRatio.AccR, width,color='green')
plt.ylabel('3lipids-in-pore', fontsize=25, color='k')
plt.xticks(lipidRatio.RepID1, tuple(['%d--%d' % (i, i+1) for i in lipidRatio.RepID1]), fontsize=10,rotation=90,color='k')
plt.yticks(fontsize=18,color='k')
plt.xticks(fontsize=9,color='k')
plt.xlim(0, 126)

print np.min(lipidRatio.AccR)
print np.max(lipidRatio.AccR)
```
